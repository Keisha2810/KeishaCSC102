```python
#Program to check if a number is even or odd
a = int(input("Insert an integer: "))
b = int(input("Insert an integer: "))
c = a + b
if c%2==0:
    print ("This is an even number, Hence even player wins")
else:
    print ("This is an odd number, Hence odd player wins")
```
